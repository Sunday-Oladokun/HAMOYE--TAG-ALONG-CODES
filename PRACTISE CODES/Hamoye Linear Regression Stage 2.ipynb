### SIMPLE LINEAR REGRESSION

A simple linear regression model estimates the relationship between two quantitative variables where one is referred to as the independent variable and the other the dependent variable. The independent variable (X) is used to predict and also called the predictor while the predicted variable is referred to as the response variable (Y) (e.g. finding the relationship between the amount of CO2 gas emitted and the number of trees cut down). The value of Y can be obtained from X by finding the line of best fit (regression line) with minimum error for the data points on a scatter plot for both variables. A simple linear regression can be represented as:

    y = ax + b

y = dependent variable

a = slope

x = independent variable

b = intercept

### ASSUMPTIONS OF LINEAR REGRESSION

Linear relationship: This assumes that the relationship between the dependent and independent variables are linear such that a straight line is the best line of fit.

Homoscedasticity, in the context of statistics and regression analysis, refers to the assumption that the variance of the residuals (the differences between observed and predicted values in a regression model) is constant across all levels of the independent variables. 

Here's a breakdown of the concept:

1. **Residuals**: In regression analysis, residuals are the differences between the observed values of the dependent variable and the values predicted by the regression model. These residuals represent the errors of the model in predicting the dependent variable.

2. **Variance of Residuals**: Homoscedasticity implies that the spread or dispersion of these residuals (i.e., their variance) remains the same regardless of the values of the independent variables. Essentially, this means that the variability of the errors is consistent across the entire range of the independent variables.

3. **Why it Matters**: Homoscedasticity is important for the validity of regression models. If the assumption of homoscedasticity is violated (i.e., the variance of residuals changes with different values of the independent variables), it can lead to biased estimates of the regression coefficients and incorrect conclusions about the significance of the predictors.

4. **Identifying Homoscedasticity**: To check for homoscedasticity, you can visually inspect plots of residuals against predicted values or against each independent variable. In a homoscedastic scenario, these plots would exhibit a random scatter around zero without any discernible pattern. On the other hand, heteroscedasticity (the violation of homoscedasticity) would show a noticeable pattern, such as the spread of residuals increasing or decreasing as predicted values change.

5. **Dealing with Violations**: If heteroscedasticity is detected, it might be necessary to use techniques like transforming the dependent or independent variables, using weighted least squares regression, or employing robust standard errors to address the issue and ensure the reliability of the regression results.

Residuals: Residuals are calculated by subtracting the predicted values from the observed values. Each residual represents how far off the model's prediction was for a specific data point.
If a data point has a positive residual, it means the observed value is higher than what the model predicted.
If a data point has a negative residual, it means the observed value is lower than what the model predicted.

Normality: the residuals are of a normal distribution. If we plot the distribution of the residuals, they yield a normal distribution

Multicolinearity: In multiple linear regression where there are more predictors, it is assumed that these variables are independent of each other without any strong correlation between them.In simple terms, this means that the independent variables must not be correlated. If they are, it would make our model complex and it undermines the statistical significance of an independent variables.

A scatter plot or heatmap can be used to visualize this.



### COST FUNCTION
A cost function is a measure of the performance of a model i.e. how far or close the predicted values are to the real values. The objective is to minimise the cost function in order for the model to continuously learn to obtain better results.

Gradient descent: an optimization algorithm that minimizes a cost function by specifying the direction to move towards to obtain a local or global minima. This is done by initially starting with random values then iteratively updating the values until the minimum cost is obtained.  A learning rate is usually chosen to determine the step size to be taken for each iteration. It is important to carefully select this parameter because, if a small step is chosen, it will take a long time to converge to the minimum cost while if too large, it can result in an overshoot surpassing the location of the minimum cost.

The general form of a polynomial regression model with degree \( n \) is:
\[ y = \beta_0 + \beta_1 x + \beta_2 x^2 + \ldots + \beta_n x^n + \epsilon \]
where \( y \) is the dependent variable, \( x \) is the independent variable, \( \beta_0, \beta_1, \ldots, \beta_n \) are the coefficients to be estimated, \( x^2, x^3, \ldots, x^n \) represent the higher-order terms, and \( \epsilon \) is the error term.

Key points about polynomial regression:

1. **Flexibility**: Polynomial regression can capture nonlinear relationships between \( x \) and \( y \). By using higher-degree polynomials, the model can fit more complex patterns in the data.

2. **Degree Selection**: The choice of the degree \( n \) of the polynomial is crucial. Higher degrees can lead to overfitting, where the model fits the noise in the data rather than the underlying relationship. Regularization techniques can be used to prevent overfitting. Higher degrees lead to more complexity.

3. **Interpretation**: Interpretation of the coefficients in polynomial regression becomes more complex with higher degrees. Each coefficient represents the effect of the corresponding degree of the independent variable \( x \).

4. **Evaluation**: To evaluate the goodness of fit of a polynomial regression model, metrics such as \( R^2 \) (coefficient of determination) and adjusted \( R^2 \) can be used. Additionally, cross-validation can help assess the model's performance on unseen data.

5. **Visualization**: Plotting the fitted polynomial curve along with the data points can provide insights into how well the model captures the underlying relationship.

6. **Practical Use**: Polynomial regression is used when the relationship between variables is suspected to be nonlinear and other regression methods (like linear regression) are insufficient.

Overall, while polynomial regression offers increased flexibility in modeling nonlinear relationships, careful consideration must be given to the choice of degree and model evaluation to ensure robust and accurate results.

### MEASURING MODEL PERFORMANCE
How well a regression model performs can be obtained by how close the predicted value is to the ground truth. It is very important to use the appropriate metric to evaluate the performance.

MEAN ABSOLUTE ERROR: MAE  is easy and intuitive such that it calculates the sum of the  average of the absolute error between the predicted values and the true values. Since the absolute difference is taken, this metric does not consider direction(OVERFITTING OR UNDERFITTING). However, because the absolute difference is obtained, it is unable to give information about the model overshooting or undershooting. The smaller the MAE is, the better the model. Therefore, if the MAE is 0, the model is perfect and accurately predicts results which is almost impossible.  The mean absolute error is more robust to outliers

```python
#Firstly, we normalise our dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression

scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
features_df = normalised_df.drop(columns=['Heating_Load', 'Cooling_Load'])
heating_target = normalised_df['Heating_Load']

#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target,
test_size=0.3, random_state=1)
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)

#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

NOTE:

x_train = independent variables(predictor) that would be used to train the model. Each row of x_train corresponds to a training example, and each column represents a feature (e.g., age, income, etc.).

y_train = the corresponding dependent varibale for independent variables. It contains the actual values or labels that the model is trained to predict based on the input features in x_train.

x_test =  Similarly to x_train, x_test contains the input features of the testing dataset. These are unseen examples that the trained model will use to make predictions.

y_test = This is the corresponding target variable for the x_test data. y_test contains the actual values or labels for the x_test examples, which are used to evaluate the performance of the trained model. In summary, the y_test are the true values.

In summary, x_train and y_train are used to train (fit) a machine learning model, while x_test and y_test are used to evaluate the performance of the trained model on unseen data. The goal is to build a model that can generalize well to new, unseen examples beyond the training data.

RESIDUL SUM OF SQUARES: Also known as the sum of squared residuals (SSR), this metric explains the variance in the representation of the dataset by the model; it measures how well the model approximates the data. A residual is the estimated error made by a model. In simpler terms, it is the difference between the nth true value and the nth predicted value by the model. RSS is the sum of the square of errors between the residuals in a model. The lower the RSS, the better the model’s estimations and vice versa.

```python
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

R SQUARED: Also known as the coefficient of determination, r-squared is a metric used in regression to determine the goodness of fit of the model. With values ranging from 0 to 1, It gives information on the percentage of the response variable  explained by the model. Mostly, the higher the value, the better the model however, this is not necessarily always true.

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3)
```

ROOT MEAN SQUARE ERROR: The Root Mean Square Error (RMSE) is another commonly used metric for evaluating the performance of regression models, including linear regression. RMSE is a measure of the average magnitude of the errors between predicted values and actual values. It is calculated by taking the square root of the mean of the squared differences between predicted and actual values.

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

Model complexity, Underfitting and Overfitting

Model complexity refers to the number of input features used to train a model and the algorithmic learning complexity. An overly complex model can be difficult to interpret, prone to overfitting and also require more computing. When creating models, it is imperative for the model to generalise well enough to make reasonable predictions on new and unseen data. An overfit model will perform well on the training data and poorly on unseen data. While a model is required to learn the actual relationship of the variables in the training set, an overfit model memorises the training set, fits the noise, outliers and irrelevant information, then makes predictions based on this noise which is incorrect. On the other hand, when a model is too simple, it can be as a result of having very few features not sufficient enough to learn details and relationships in the data.  In a later section, we will discuss methods that can be used to achieve optimal and acceptable model complexities while avoiding overfitting and underfitting. 

The Bias-Variance tradeoff

Bias and variance are common occurrences in machine learning and there is a constant struggle to achieve low bias and variance. Bias is a measure of correctness of a model i.e. how far off is a model from being correct? While high bias results in an increase in the error by making assumptions which prevent the model from capturing relevant relationships between the predictors and response variable, low bias gives lower error and also prevents underfitting by capturing important relationships. On the other hand, variance tells how much the values estimated by a model will vary across different training data. When the variance is low, it means that there is only a small change in the estimate of the model with  new training  data. A high variance causes overfitting such that the changes in estimates obtained with new training data is large because the model is so complex that it has now learnt patterns from one training data such that it cannot generalise to other training sets. While it is essential to obtain low bias and low variance, it is almost impossible to achieve this simultaneously which is where the ‘bias-variance tradeoff’ occurs. 

### PENALIZATION METHODS

Penalization methods in linear regression are techniques used to impose additional constraints or penalties on the model coefficients during the training process. These methods are primarily employed to prevent overfitting and to encourage the model to generalize well to new, unseen data.

Regularization is a method used to make complex models simpler by penalising coefficients to reduce their magnitude, variance in the training set and in turn, reduce overfitting in the model. Regularization occurs by shrinking the coefficients in the model towards zero such that the complexity term added to the model will result in a bigger loss for models with a higher complexity . There are two types of regression techniques such as Ridge and Lasso regression.

Variance in machine learning: In the context of machine learning and data analysis, the term "variance" refers to the variability or spread of the data points within a dataset. Specifically, when discussing variance in the training set, we are interested in understanding the distribution and dispersion of the feature values (input variables) within the samples used for training a machine learning model.

Here are some key points related to variance in the training set:

1. **Feature Variance**:
   - For each feature (or variable) in the training set, variance measures how much the values of that feature vary or spread out around their mean (average).
   - Features with low variance indicate that their values are relatively similar or constant across the training samples.
   - Features with high variance indicate that their values are widely spread out and exhibit more variability across the training samples.

2. **Impact on Model Training**:
   - Features with low variance may not provide much useful information for training the model, as they do not exhibit significant variation that can help discriminate between different classes or outcomes.
   - Features with high variance can sometimes lead to overfitting, especially if the model tries to learn from noise or irrelevant patterns present in the training data.

3. **Identifying Low-Variance Features**:
   - It's important to identify and potentially remove features with very low variance (close to zero) from the training set, as they are unlikely to contribute meaningfully to the model's predictive ability.
   - Removing low-variance features can help simplify the model and reduce the risk of overfitting.

4. **Assessing Data Quality**:
   - Variance in the training set can also be a useful metric for assessing the quality and diversity of the dataset.
   - A wide range of variance across different features can indicate that the dataset captures a diverse set of patterns and characteristics, which can be beneficial for training robust and generalizable models.

In practice, you can calculate the variance of features in the training set using statistical functions or libraries in Python such as `numpy.var()` or `pandas.DataFrame.var()`. This can help you analyze the distribution of feature values and make informed decisions about feature selection, preprocessing, and model training to optimize performance and avoid common pitfalls like overfitting.
