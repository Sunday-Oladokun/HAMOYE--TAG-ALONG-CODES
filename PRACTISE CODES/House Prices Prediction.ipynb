### DATA PREPROCESSING

```python
# Check for missing values
df.isna().sum()
```

```python
df.columns
```

```python
# Encoding categorical variables
categorical_variables = ['waterfront', 'view', 'condition']

# Create an instance of label encoder
label_encoder = LabelEncoder()

# For loop for encoding labels
for category in categorical_variables:
    df[category] = label_encoder.fit_transform(df[category])
```

```python
# Check to confirm the encoding made
df.head()
```

```python
# Split the dataset into target and features

# depend variable
y = df['price']

# independent variables
x = df.drop(['id', 'date', 'price', 'lat', 'long'], axis=1)
```

```python
# Check y
y.head()
```

```python
# Check x
x.head()
```

```python
# Split the features and targets into training and test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

### MODEL TRAINING

```python
# Fit the model to the training data

# Select model
model = LinearRegression()

# Fit Model
model.fit(x_train,y_train)
```

### MODEL EVALUATION

```python
# Predict the target values for the features dataset
y_prediction = model.predict(x_test)
```

```python
# Convert the predicted values to a dataframe and merge it to the y_test
predicted_price = pd.DataFrame(y_prediction, columns=['predicted price'])
features_and_prediction = x_test.merge(predicted_price,how='inner', left_index=True, right_index=True)
```

```python
# view our data to see the predicted prices
features_and_prediction.head()
```

```python
# Append the y_test prices to the dataframe above to see the original vs predicted price

# Convert y_test to a dataframe and append to the dataframe
original_price = pd.DataFrame(y_prediction, columns=['original price'])
original_predicted_prices_df = features_and_prediction.merge(y_test,how='inner', left_index=True, right_index=True)
```

From the above, we can see that the model is not close to predicting the prices accurately and we will confirm this when we evaluate our model 

```python
# Evaluate the model using mean square error
mse = mean_squared_error(y_test, y_prediction)
mse
```

```python
# Evaluate the model using r_squared
r2_score = r2_score(y_test, y_prediction)
round(r2_score, 3)
```
